```python
import pandas as pd
import numpy as np
```

```python
# LOAD THE DATA SET
dataset=pd.read_excel('Drilling Datasets for Mud-loss Prediction.xlsx')
dataset
```

```python
df=dataset.copy()
df
```

```python
df=df.dropna()
df
```

```python
#ROUNDING UP THE DATASET TO 2 DECIMAL PLACES
df=np.round(df, decimals=2)
df.head()
```

```python
#PREPARE THE OUTPUT BY CREATING ANOTHER COLUMN (Mud_loss) AND FILLING IT WITH ZEROS.
df['Mud_loss']=0
df
```

```python
for x in df.index:
    if df.loc[x,'Mud loss (bbl/hr)'] >0.1:
        df.loc[x,'Mud_loss']=1
```

```python
#SHUFFLE THE DATA
df=df.sample(frac=1)
df
```

```python
index_20percent=int(0.2*len(df.iloc[:,0]))
index_20percent
```

```python
Xtest=df.iloc[:index_20percent,:-1]
Ytest=df.iloc[:index_20percent,-1]

Xtrain=df.iloc[index_20percent:,:-1]
Ytrain=df.iloc[index_20percent:,-1]
```

```python
Xtest=df.iloc[:index_20percent,:-1]
Ytest=df.iloc[:index_20percent,-1]

Xtrain=df.iloc[index_20percent:,:-1]
Ytrain=df.iloc[index_20percent:,-1]
```

```python
#NORMALIZING THE DATA USING STANDARDIZATION

mean=Xtrain.mean(axis=0)
Xtrain-=mean
std=Xtrain.std(axis=0)
Xtrain/=std

Xtest-=mean
Xtest/=std
```

```python
#THE DATA SET HAS BEEN NORMALIZED
print(Xtrain.shape)
print(Ytrain.shape)
print(Xtest.shape)
print(Ytest.shape)
print(Xtrain[0:3])
print(Ytrain[0:3])
print(Xtest[0:3])
print(Ytest[0:3])
```

```python
#CREATING A NEURAL NETWORK MODEL
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model=keras.Sequential(
[keras.layers.Dense(units=256,input_dim=Xtrain.shape[1],activation='relu',name='input_layer'),
keras.layers.Dense(units=128,input_dim=Xtrain.shape[1],activation='relu',name='layer2'),
#keras.layers.Dropout(rate=0.9),
keras.layers.Dense(units=64,input_dim=Xtrain.shape[1],activation='relu',name='layer3'),
#keras.layers.Dropout(rate=0.9),
keras.layers.Dense(units=32,input_dim=Xtrain.shape[1],activation='relu',name='layer4'),
keras.layers.Dense(units=1,activation='sigmoid',name='output_layer')]      
)
```

```python
#  COMPILING THE MODEL
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
# TRAINING THE MODEL
history=model.fit(Xtrain,Ytrain,epochs=10,verbose=1)
```

```python
model.summary()
```

```python
#STEP 13: EVALUATING THE DATA
loss,accuracy=model.evaluate(Xtrain,Ytrain)
print('loss= ',loss)
print('accuracy= ',accuracy)
```

```python
print(history.params)
```

```python
scores=model.evaluate(Xtrain,Ytrain)
```

```python
#STEP 14: PREDICTING THE MODEL
prediction=model.predict(Xtest)
```

```python
#PREDICTION THE MUD LOSS FOR THE FIRST 10 ROWS
print(prediction[3].round())
```

```python
Ytest.iloc[3]
```

```python
#EVALUATION OF THE TESTING SET
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
accuracy=accuracy_score(Ytest,prediction.round())
precision=precision_score(Ytest,prediction.round())
recall=recall_score(Ytest,prediction.round())
f1score=f1_score(Ytest,prediction.round())

print('Accuracy:%.2f%%' %(accuracy*100.0))
print('Precision:%.2f%%' %(precision*100.0))
print('Recall:%.2f%%' %(recall*100.0))
print('F1-score:%.2f%%' %(f1score*100.0))
```
